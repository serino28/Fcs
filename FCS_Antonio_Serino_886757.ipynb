## Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates.

Per poter effettuare la query numero 12 è stato necessario innanzitutto pulire il dataset, rimuovendo tutte quelle righe che contenevano un valore nullo per la descrizione.
Successivamente vi è un preprocessing per uniformare i testi e pulili da caratteri inutili ai fini del calcolo (link, punteggiatura, singoli caratteri particolari e minuscolo).

Il dataset viene inoltre ordinato in ordine crescente sulla base del proprio id.

Per poter ritrovare tutte le coppie di cani che condividono lo stesso sesso e la stessa razza viene effettuata una self join sul dataframe stesso sulle colonne "breed_primary", "breed_secondary" e "sex": il risultato sarà un dataframe con 7 colonne che descriveranno id del primo cane, descrizione del primo cane, id secondo cane, id secondo cane, sesso e razza primaria e secondaria. Su questo dataframe si specifica la condizione per cui il primo id deve sempre essere diverso dal secondo id ed inoltre il primo id deve essere minore del secondo, così da non avere righe duplicate con gli id invertiti.

Dunque si effettua il calcolo della similarità che per ogni riga calcola la percentuale di parole in comune come l'intersezione delle due parole fratto l'unione. Viene successivamente filtrato il dataframe andando ad ottenere quello definitivo con la percentuale superiore o pari a 0.90.

```python
df1['description'].isnull().sum()
```

```python
df1 = df1.dropna(subset=['description'])#rimuovo le descrizioni nulle
```

```python
df1['description'].isnull().sum() 
```

```python
import nltk
from nltk.tokenize import word_tokenize
import string
import re

# Scarichiamo i dati necessari per la pulizia
nltk.download('punkt')
def morandi(text):
    """
    Pulisce il testo delle descrizioni.
    """
    # Rimuoviamo i link
    text = re.sub(r'http\S+', '', text)
    
    # Rimuove la punteggiatura
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Rimuoviamo le singole lettere speciali
    text = re.sub(r'\b\w\b', '', text)
    
    # Rimuoviamo i caratteri indesiderati
    text = re.sub(r'[^\w\s]', '', text)
    
    # Rimuoviamo i caratteri di tabulazione e a capo
    text = re.sub(r'[\n\t]', '', text)
    
    # testo in minuscolo
    text = text.lower()

    return text
```

```python
df1['description'] = df1['description'].apply(morandi) #pulisco il testo da caratteri che lo appesantiscono (no rimozione stopword)
```

#### similarità: intersezione delle parole tra le due descrizioni / unione delle parole delle due descrizioni (tempo: 17 minuti)

```python
import time
```

```python
def percentuale_parole_comuni(str_1, str_2):
    # dividi le stringhe in parole
    parole_1 = set(str_1.split())
    parole_2 = set(str_2.split())
    # calcola la percentuale di parole comuni
    percentuale = len(parole_1.intersection(parole_2)) / len(parole_1.union(parole_2))
    return percentuale
#semplice calcolo che si basa sul rapporto tra l'intersezione e l'unione di due parole (più o meno un'ora per il completamento)
#ricorda che si è spento il pc
```

```python
start_time = time.time()

result['percentuale_parole_comuni'] = result.apply(lambda row: percentuale_parole_comuni(str(row['description_x']), str(row['description_y'])), axis=1)

end_time = time.time()

print("Tempo di esecuzione:", end_time - start_time, "secondi")
```
